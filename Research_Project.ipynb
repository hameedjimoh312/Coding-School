```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/frontiertechinstitute/datasets/refs/heads/main/Passwords/sampled_passwords.csv"

df = pd.read_csv(url)
```

```python

df.head()
```

```python

duplicates_df = df[df.duplicated(subset = "Password")]
duplicates_df['Strength_Level'].describe()

```

```python
sample = duplicates_df.head(1000)


counts = sample['Strength_Level'].value_counts()
x = counts.index
y = counts.values

plt.bar(x,y)
plt.xticks([0,1,2,3,4])


plt.xlabel("Strength Number")
plt.ylabel("Frequency")
plt.title("Duplicated value strengths and frequency")


plt.show()
```
